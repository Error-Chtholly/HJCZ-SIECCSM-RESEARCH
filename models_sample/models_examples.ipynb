```python
# K-means聚类示意图
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# 随机生成数据，确保簇分布均匀
X1 = np.random.normal(1, 0.5, (100, 2))  # 第一簇，中心在(1, 1)
X2 = np.random.normal(2, 0.5, (100, 2))  # 第二簇，中心在(2, 3)
X3 = np.random.normal(3, 0.5, (100, 2))  # 第三簇，中心在(3, 1)
X4 = np.random.normal(4, 0.5, (100, 2))  # 第四簇，中心在(4, 4)
X5 = np.random.normal(5, 0.5, (100, 2))  # 第五簇，中心在(5, 2)

# 合并成一个数据集
X = np.vstack([X1, X2, X3, X4, X5])

# K-means聚类
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# 绘制散点图，使用鲜艳的颜色
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='plasma', marker='o')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='x', label='Centroids')

# 加粗图框
plt.gca().spines['top'].set_linewidth(4)
plt.gca().spines['right'].set_linewidth(4)
plt.gca().spines['left'].set_linewidth(4)
plt.gca().spines['bottom'].set_linewidth(4)

# 删除网格线、X/Y轴标题和图题
plt.xticks([])
plt.yticks([])
plt.gca().set_facecolor('white')

# 保存图像为600 dpi
plt.savefig("K.png", dpi=600)

# 显示图像
plt.show()
```

```python
# 可达性分布示意图
import numpy as np
import matplotlib.pyplot as plt

# 设置字体为宋体
plt.rcParams['font.family'] = 'SimSun'

# 模拟127个数值点，数据分布为高低起伏式
np.random.seed(42)

# 每个区间的范围和数据分布
x1 = np.random.uniform(1, 1.8, 18)  # 第一部分：高
x2 = np.random.uniform(2.2, 2.8, 18)  # 第二部分：低
x3 = np.random.uniform(3.2, 3.8, 18)  # 第三部分：高
x4 = np.random.uniform(1.3, 1.9, 18)  # 第四部分：低
x5 = np.random.uniform(2.5, 3.1, 18)  # 第五部分：低
x6 = np.random.uniform(3.9, 4.5, 18)  # 第六部分：高
x7 = np.random.uniform(2.7, 3.3, 17)  # 第七部分：低

# 合并数据
x = np.concatenate([x1, x2, x3, x4, x5, x6, x7])

# 为y值设置相似的分布，增加高低起伏
y1 = np.random.uniform(4, 4.7, 18)
y2 = np.random.uniform(1.0, 1.7, 18)
y3 = np.random.uniform(4.3, 5, 18)
y4 = np.random.uniform(1.4, 2.0, 18)
y5 = np.random.uniform(1.6, 2.4, 18)
y6 = np.random.uniform(3.9, 4.6, 18)
y7 = np.random.uniform(1.5, 2.2, 17)

# 合并y数据
y = np.concatenate([y1, y2, y3, y4, y5, y6, y7])

# 绘制散点图
plt.figure(figsize=(8, 4))
plt.scatter(x, y, c='red', marker='o')

# 设置坐标轴加粗
plt.gca().spines['top'].set_linewidth(2)
plt.gca().spines['right'].set_linewidth(2)
plt.gca().spines['left'].set_linewidth(2)
plt.gca().spines['bottom'].set_linewidth(2)

# 只显示y轴刻度
plt.xticks([])  # 隐藏x轴刻度标签
plt.yticks(np.arange(1, 6, 1), fontsize=21)  # 设置y轴刻度标签为1到5之间的整数

# 仅显示左侧y轴和下侧x轴的坐标轴
plt.gca().spines['top'].set_color('none')
plt.gca().spines['right'].set_color('none')

# 多项式拟合（4阶拟合）
coefficients = np.polyfit(x, y, 4)  # 4阶多项式拟合
poly = np.poly1d(coefficients)  # 创建多项式对象

# 生成拟合曲线
x_fit = np.linspace(min(x), max(x), 100)  # 创建更细的x点以绘制平滑曲线
y_fit = poly(x_fit)

# 绘制蓝色虚线拟合曲线
plt.plot(x_fit, y_fit, 'b--')

# 保存图像为600 dpi
plt.savefig("scatter_plot_with_fit_curve_600dpi.png", dpi=600)

# 显示图像
plt.show()
```

```python
# SHAP示意图
import shap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.ensemble import RandomForestRegressor

# 设置字体为宋体
plt.rcParams['font.family'] = 'SimSun'
plt.rcParams['font.size'] = 21  # 设置整体文字大小

# 生成更多模拟数据
X, y = make_regression(n_samples=500, n_features=5, noise=0.1, random_state=42)

# 创建一个回归模型并训练
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X, y)

# 创建SHAP解释器
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X)

# 绘制SHAP Summary Plot
plt.figure(figsize=(10, 8))
shap.summary_plot(shap_values, X, plot_type="dot")
plt.savefig("shap_summary_plot_500points.png", dpi=600)
plt.show()

# 绘制SHAP Bar Plot
plt.figure(figsize=(8, 6))
shap.summary_plot(shap_values, X, plot_type="bar")
plt.savefig("shap_bar_plot_500points.png", dpi=600)
plt.show()

# 绘制SHAP Force Plot（针对一个实例）
shap.initjs()
plt.figure(figsize=(8, 6))
shap.force_plot(explainer.expected_value, shap_values[0], X[0], matplotlib=True)
plt.savefig("shap_force_plot_500points.png", dpi=600)
plt.show()
```

```python
import shap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.ensemble import RandomForestRegressor

# 设置字体为宋体
plt.rcParams['font.family'] = 'SimSun'
plt.rcParams['font.size'] = 14  # 设置整体文字大小

# 生成模拟数据
X, y = make_regression(n_samples=500, n_features=5, noise=0.1, random_state=42)

# 创建中文特征名称
feature_names = ['特征1', '特征2', '特征3', '特征4', '特征5']
X_df = pd.DataFrame(X, columns=feature_names)

# 创建回归模型并训练
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_df, y)

# 创建SHAP解释器
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_df)

# 绘制SHAP Summary Plot
plt.figure(figsize=(10, 8))
shap.summary_plot(shap_values, X_df, plot_type="dot", 
                 show=False, color_bar_label='特征值')
plt.xlabel("SHAP值（因子对模型输出的解释力）", fontsize=14)
plt.gcf().axes[-1].set_ylabel('特征值', fontsize=14)
plt.gca().set_yticklabels(feature_names)

# 修改渐变条带的High和Low为中文
color_bar = plt.gcf().axes[-1]
color_bar.set_yticklabels(['低', '高'])  # 修改这里

plt.savefig("shap_summary_plot_500points.png", dpi=600, bbox_inches='tight')
plt.show()

# 绘制SHAP Bar Plot
plt.figure(figsize=(8, 6))
shap.summary_plot(shap_values, X_df, plot_type="bar", show=False)
plt.xlabel("平均绝对SHAP值（特征重要性）", fontsize=21)
plt.savefig("shap_bar_plot_500points.png", dpi=600, bbox_inches='tight')
plt.show()

# 绘制SHAP Force Plot（针对一个实例）
shap.initjs()
plt.figure(figsize=(12, 6))
shap.force_plot(explainer.expected_value, 
               shap_values[0], 
               X_df.iloc[0], 
               matplotlib=True,
               feature_names=feature_names,
               text_rotation=15)
plt.xlabel("SHAP值（因子对模型输出的作用力）", fontsize=21)
plt.savefig("shap_force_plot_500points.png", dpi=600, bbox_inches='tight')
plt.show()
```
